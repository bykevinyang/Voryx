```python
import numpy as np
from numpy import arccos
from scipy import spatial
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity
from collections import OrderedDict
```

```python
embeddings_dict = {}
with open(GloVe_path, 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        token = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[token] = vector
```

```python
# O(n)
def embed_sentence(sentence):
  words = sentence.split()
  embed = []
  for word in words:
    embed.append(embeddings_dict[word][1])
  return embed
```

```python
# O(1)
def cosine_similarity(embed1, embed2):
  norm1 = length_normalize(embed1)
  norm2 = length_normalize(embed2)
  return spatial.distance.cosine(norm1, norm2)
```

```python
# O(n)
def length_normalize(words, size = 15):
  length = len(words)
  output = words.copy()

  if length < size:
    difference = size - length
    for i in range(difference):
      output.append(0)
    return tuple(output)
  elif length > size:
      raise Exception("Input length is greater than size. Size is: " + str(size))
  else:
    return tuple(output)
```
